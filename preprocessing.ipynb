```python
import csv
import itertools
```

```python
read_link='english_condensedStats.csv'

clean_link='without_duplicates_condensedStats.csv'

# Removing duplicate rows from the data set
with open(read_link, 'r') as in_file, open(clean_link, 'w') as out_file:
    seen = set()
    for line in in_file:
        if line in seen: continue
            
        seen.add(line)
        out_file.write(line)
```

```python
link='without_duplicates_condensedStats.csv'

with open(link, newline='', encoding='utf8') as f:
    lines = itertools.islice(f,1,None)
    reader = csv.reader(lines)
    
    cnt={}
    for row in reader:
        # Including only videos which at least one tag
        if(row[6]!=''):
                check_arr=row[6].split(",")
                for i in range(0,len(check_arr)):
                    cnt[check_arr[i]]=0
```

```python
# Dictionary of occurrences of each tag in the data set
with open(link, newline='', encoding='utf8') as f:
    lines = itertools.islice(f,1,None)
    reader = csv.reader(lines)
    
    for row in reader:
            if(row[6]!=''):
                    check_arr=row[6].split(",")
                    for i in range(0,len(check_arr)):
                        cnt[check_arr[i]]+=1

    print(cnt)
```
